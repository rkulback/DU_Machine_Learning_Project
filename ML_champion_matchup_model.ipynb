```python
# PRE MATCH PREDICTING MODEL #
#models for predicting game outcomes;

# Dependencies
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Store filepath in a variable
path = "champion_matchups_2019_table.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
master_df = pd.read_csv(path, encoding="ISO-8859-1")
```

```python
master_df.columns
```

```python
master_df.head()
```

```python
master_df.replace('blue loss', 0, inplace=True)
master_df.replace('blue win', 1, inplace=True)
```

```python
X = master_df.drop(['game_id', 'jg_result', 'mid_result', 'bot_result',
                   'sup_result', 'jg_matchup', 'top_matchup', 'bot_matchup', 
                   'mid_matchup', 'sup_matchup', 'top_result'], axis=1)
y = master_df['top_result'].values.reshape(-1, 1)

class_names = y

X.columns
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)

y_pred = classifier.fit(X_train, y_train).predict(X_test)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, classifier.predict(X_test))
```

```python
## try to build this function to use to study the datasets

def my_champions():
    
    blue_jungle_champ = input("enter the blue jungle side champion: ")
    red_jungle_champ = input("enter the red jungle side champion: ")
    blue_top_champ = input("enter the blue side top champion: ")
    red_top_champ = input("enter the red side top champion: ")
    blue_mid_champ = input("enter the blue side mid champion: ")
    red_mid_champ = input("enter the red side mid champion: ")
    blue_bot_champ = input("enter the blue side bot champion: ")
    red_bot_champ = input("enter the red side bot champion: ")
    blue_sup_champ = input("enter the blue side support champion: ")
    red_sup_champ = input("enter the red side support champion: ")
    
    connector = 'stack_here'
    
    ## pull values from dataframe
    
    blue_jungle = stats_book_a.loc[stats_book_a['winner'] == blue_jungle_champ]
    red_jungle = blue_jungle.loc[blue_jungle['loser'] == red_jungle_champ]
  
    red_jungle['matchup'] = red_jungle.winner.astype(str).str.cat(red_jungle.loser.astype(str), sep=" , ")
    master_jungle_df = red_jungle.drop(columns=['winner', 'loser', 'win', 'loss'])
    master_merged_jungle = pd.merge(master_jungle_df, stats_book_b, on='matchup')
    final_jungle = master_merged_jungle[['kills', 'assists', 'deaths', 'csat15', 'oppcsat15', 
                                   'total_gold', 'damage', 'matchup_strength']]
    final_jungle['stack'] = connector
    
    
    blue_top = stats_book_a.loc[stats_book_a['winner'] == blue_top_champ]
    red_top = blue_top.loc[blue_top['loser'] == red_top_champ] 
#     red_top['stack'] = connector
    red_top['matchup'] = red_top.winner.astype(str).str.cat(red_top.loser.astype(str), sep=" , ")
    master_top_df = red_top.drop(columns=['winner', 'loser', 'win', 'loss'])
    master_merged_top = pd.merge(master_top_df, stats_book_b, on='matchup')
    final_top = master_merged_top[['kills', 'assists', 'deaths', 'csat15', 'oppcsat15', 
                                   'total_gold', 'damage', 'matchup_strength']]
    
    
    blue_mid = stats_book_a.loc[stats_book_a['winner'] == blue_mid_champ]
    red_mid = blue_mid.loc[blue_mid['loser'] == red_mid_champ] 
#     red_mid['stack'] = connector
    red_mid['matchup'] = red_mid.winner.astype(str).str.cat(red_mid.loser.astype(str), sep=" , ")
    master_mid_df = red_mid.drop(columns=['winner', 'loser', 'win', 'loss'])
    master_merged_mid = pd.merge(master_mid_df, stats_book_b, on='matchup')
    final_mid = master_merged_mid[['kills', 'assists', 'deaths', 'csat15', 'oppcsat15', 
                                   'total_gold', 'damage', 'matchup_strength']]
    
    blue_bot = stats_book_a.loc[stats_book_a['winner'] == blue_bot_champ]
    red_bot = blue_bot.loc[blue_bot['loser'] == red_bot_champ] 
#     red_bot['stack'] = connector
    red_bot['matchup'] = red_bot.winner.astype(str).str.cat(red_bot.loser.astype(str), sep=" , ")
    master_bot_df = red_bot.drop(columns=['winner', 'loser', 'win', 'loss'])
    master_merged_bot = pd.merge(master_bot_df, stats_book_b, on='matchup')
    final_bot = master_merged_bot[['kills', 'assists', 'deaths', 'csat15', 'oppcsat15', 
                                   'total_gold', 'damage', 'matchup_strength']]
    
    blue_sup = stats_book_a.loc[stats_book_a['winner'] == blue_sup_champ]
    red_sup = blue_sup.loc[blue_sup['loser'] == red_sup_champ] 
#     red_sup['stack'] = connector
    red_sup['matchup'] = red_sup.winner.astype(str).str.cat(red_sup.loser.astype(str), sep=" , ")
    master_sup_df = red_sup.drop(columns=['winner', 'loser', 'win', 'loss'])
    master_merged_sup = pd.merge(master_sup_df, stats_book_b, on='matchup')
    final_sup = master_merged_sup[['kills', 'assists', 'deaths', 'csat15', 'oppcsat15', 
                                   'total_gold', 'damage', 'matchup_strength']]
 
    final_top['stack'] = connector
    final_mid['stack'] = connector
    final_bot['stack'] = connector
    final_sup['stack'] = connector
    #merge the datapoints
    
    merge_1 = pd.merge(final_jungle, final_top, on='stack')
    
    merge_2 = pd.merge(merge_1, final_mid, on='stack')
    
    merge_3 = pd.merge(merge_2, final_bot, on='stack')
    
    final_df = pd.merge(merge_3, final_sup, on='stack')
    
    real_X = final_df.drop(['stack'], axis=1)
    
    return classifier.predict(real_X)
            


```
